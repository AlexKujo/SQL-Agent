```python
def truncate_embedding(embedding, target_dim):
    truncated = np.array(embedding[:target_dim])
    # truncated = normalize(truncated)
    return truncated.tolist()


def normalize(vector):
    norm = np.linalg.norm(vector)
    if norm > 0:
        vector = vector / norm
    return vector
```

```python
embeddings = OllamaEmbeddings(model="embeddinggemma", base_url="http://localhost:11434")
```

```python
test_text = "тестовый текст для проверки размерности"
dimension = 512

vector = embeddings.embed_query(test_text)

trunk_vector = truncate_embedding(vector, dimension)
```
